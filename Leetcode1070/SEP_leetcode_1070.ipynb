```python
print("Data types of columns:")
print(df_1.dtypes)
```

```python
# Define a function to calculate the rank of each year for each product_id
def calculate_year_rank(group):
    group['YEAR_RANK'] = group['year'].rank(method='min')
    return group

# Apply the function to calculate the year rank
df_1 = df_1.groupby('product_id').apply(calculate_year_rank)

print("\nDataFrame after calculating year rank:")
df_1.head(10)  # Display the DataFrame after calculating year rank
```

```python
# Filter the DataFrame to get only the rows where YEAR_RANK is 1
result = df_1[df_1['YEAR_RANK'] == 1] 

result.head(10)
```

```python
# Select the required columns
result = result[['product_id', 'year', 'quantity', 'price']]

result.head()
```

```python
import datetime
# Get the current date
current_date = datetime.datetime.now()

# Format the current date as dd_mm_yy
date_string = current_date.strftime("%d_%m_%y")

# Define the file path for saving the CSV file
output_file = fr'C:\Users\akrjha\Desktop\result_leetcode_{date_string}.csv'

# Save the DataFrame as a CSV file
result.to_csv(output_file, index=False)

print("CSV file saved successfully at:", output_file)
```
